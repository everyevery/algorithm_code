# 7.1 Lazy Evaluation

    sqr (sqr (3+4))

위 expression에 대해서 아래와 같이 reduction 될 것이라 생각하지만 이것은 lazy evaluation이 아니다. 3+4와 7*7에 대한 evaluation 횟수가 동일하다.

    sqr (sqr (3+4))
    = sqr (3+4) * sqr (3+4)
    = ((3+4)*(3+4)) * ((3+4)*(3+4))
    = ...
    = 2401

실제로는 3+4, 7*7이 1번씩만 계산된다.

    sqr (sqr (3+4))
    = let x = sqr (3+4) in x*x
    = let y = 3+4 in
      let x = y*y in x*x
    = let y = 7 in
      let x = y*y in x*x
    = let x = 49 in x*x
    = 2401

'Lazy evaluation arguments는 필요한 시점에 단 한 번만 evaluate 된다.' 결론?

위 정의로는 부족한 것이 있다.

    sqr (head xs)

위 코드를 evaluate하기 위해서는 xs 전체가 아닌 (head xs)만 evaluate하면 된다. 따라서 xs는 y:ys의 형태가 된다.
